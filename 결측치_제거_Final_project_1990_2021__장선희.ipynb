```python
import requests
import time

# 1. Ïó≠ÏßÄÏò§ÏΩîÎî© Ìï®Ïàò Ï†ïÏùò (ÏòÅÏñ¥ Î∞òÌôò, city Ïö∞ÏÑ† ‚Üí town ‚Üí village ‚Üí municipality)
def reverse_geocode(lat, lon):
    try:
        url = f"https://nominatim.openstreetmap.org/reverse?lat={lat}&lon={lon}&format=json"
        headers = {
            "User-Agent": "geo-reverse-city-bot",
            "Accept-Language": "en"  # ÏòÅÏñ¥Î°ú Î∞òÌôòÎêòÍ≤å ÏÑ§Ï†ï
        }
        response = requests.get(url, headers=headers, timeout=10)
        if response.ok and response.json():
            address = response.json().get("address", {})
            city = address.get("city") or address.get("town") or address.get("village") or address.get("municipality")
            return city
    except Exception as e:
        print(f"‚ùå Error at ({lat}, {lon}) ‚Üí {e}")
    return None

# 2. cityÍ∞Ä Unknown/Í≤∞Ï∏°Ïù¥Í≥† ÏúÑÍ≤ΩÎèÑ, provstate, countryÍ∞Ä ÏûàÎäî Ìñâ ÌïÑÌÑ∞ÎßÅ
mask_geo_city = (
    ((df_filled['city'].isnull()) | (df_filled['city'] == 'Unknown')) &
    df_filled['provstate'].notnull() &
    df_filled['country_txt'].notnull() &
    df_filled['latitude'].notnull() &
    df_filled['longitude'].notnull()
)
df_city_geo_target = df_filled[mask_geo_city].copy()
print(f"üìå Ïó≠ÏßÄÏò§ÏΩîÎî© city Î≥¥Í∞Ñ ÎåÄÏÉÅ Ìñâ Ïàò: {len(df_city_geo_target)}")

# 3. Ïó≠ÏßÄÏò§ÏΩîÎî© Ïã§Ìñâ
city_results = []
for idx, row in df_city_geo_target.iterrows():
    lat, lon = row['latitude'], row['longitude']
    city = reverse_geocode(lat, lon)
    city_results.append(city)
    print(f"[{idx}] ({lat}, {lon}) ‚Üí {city}")
    time.sleep(1)  # ÏöîÏ≤≠ Ï†úÌïú ÌîºÌïòÍ∏∞ ÏúÑÌï¥ Ïä¨Î¶Ω

# 4. Í≤∞Í≥º Î∞òÏòÅ
df_filled.loc[mask_geo_city, 'city'] = city_results
print("‚úÖ Ïó≠ÏßÄÏò§ÏΩîÎî© Í∏∞Î∞ò city Î≥¥Í∞Ñ ÏôÑÎ£å!")
```

```python
import re

def extract_place(summary):
    if pd.isnull(summary):
        return None

    # ‚ë† Í¥ÑÌò∏ Ïïà Ïû•ÏÜå Ï∂îÏ∂ú: e.g. (Baghdad)
    match = re.search(r"\((.*?)\)", summary)
    if match:
        return match.group(1).strip()

    # ‚ë° 'in XXXX' Ìå®ÌÑ¥ Ï∂îÏ∂ú
    match = re.search(r"\bin\s+([A-Z][a-zA-Z\s\-]+)", summary)
    if match:
        return match.group(1).strip()

    # ‚ë¢ 'at XXXX' Ìå®ÌÑ¥ Ï∂îÏ∂ú
    match = re.search(r"\bat\s+([A-Z][a-zA-Z\s\-]+)", summary)
    if match:
        return match.group(1).strip()

    return None
```

```python
import requests

def geocode_place(place_name):
    try:
        url = f"https://nominatim.openstreetmap.org/search?q={place_name}&format=json&limit=1"
        headers = {"User-Agent": "geo-latlon-bot"}
        response = requests.get(url, headers=headers, timeout=10)
        if response.ok and response.json():
            data = response.json()[0]
            return float(data['lat']), float(data['lon'])
    except Exception as e:
        print(f"‚ùå Geocode error for '{place_name}': {e}")
    return None, None
```
